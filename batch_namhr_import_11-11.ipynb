
# Database Connection
____  


1. Create connection to the database with SQLalchemy
2. Instantiates meta as the extracted metadata from the schema **"public"**
3. Autoreflects the tables found in public and turns them into classes.
    - The scope of which tables are reflected is defined in **"working_tables"** variable
4. Instantiates the Base class as an automapped Base (instead of a Base where tables are declared, in "declarative_base")
5. Activates the Base (with Base.prepare())
6. Creates global class names based on their respective counterparts which have been autoreflected and turned into models taht live in **Base.classes**
7. Sets the database to autocommit, and autoflush.Since we are using the <code>"sesion.merge"</code> method, this is a convenience for now. Eventually we'll make it a bit more secure.

# Populating Drug Exposure from Yes/No survey fields
- Note: For IHR, I pegged start time to diagnosis time, because the survey question was worded as such.

```python
drug_source = lambda x: ihr.data[ihr.data[x] =='1 Yes'][['person_id', x, 'dx_date', 'dxc_diagnosis', 'dxc_secondaryto']]
postmap = lambda x: getattr(getattr(ihr, x), 'postmap').loc[ihr.data[x] =='1 Yes']
```

```python
def output_drug_exp(x):
    drug_trunc = drug_staging[drug_fields]
    drug_trunc['person_id'] = drug_source(x)['person_id']
    drug_trunc['drug_concept_id'] = postmap(x).dropna()['drug_concept_id']
    drug_trunc['drug_exposure_start_date'] = drug_source(x).loc[drug_trunc['person_id'] == drug_source(x)['person_id']]['dx_date']
    drug_trunc['relevant_condition_concept_id'] = 36774942
    drug_trunc['drug_source_value'] = drug_source(x)[x].apply(lambda y: str(str(x) + " - " + str(y)))
    drug_trunc['drug_type_concept_id'] = 38000175 
    if len(drug_source(x)) == len(drug_trunc):
        return drug_trunc
    elif len(drug_source(x)) > len(drug_trunc):
        return drug_trunc
        print('Quality check failed. Fewer entries in drug_exposure than source data.')
    else:
        return ("Quality check failed CRITICALLY. More entries than people who took the drug.")
```
